## Data cleaning

### Subtask:
Clean the data by handling inconsistencies and outliers, and converting data types as needed. Focus on the 'Rent' column as the target variable.


**Reasoning**:
Examine the 'Rent' column for outliers using a box plot, handle them by capping, and check data types of specified columns, converting them to numeric types if necessary.



```python
import matplotlib.pyplot as plt

# 1. Examine 'Rent' for outliers and handle them
plt.figure(figsize=(8, 6))
plt.boxplot(df['Rent'])
plt.title('Box Plot of Rent')
plt.ylabel('Rent')
plt.show()

# Cap outliers in 'Rent'
rent_99th = df['Rent'].quantile(0.99)
df['Rent'] = df['Rent'].clip(upper=rent_99th)

# 2. Check and convert data types
for col in ['Rent', 'Size', 'BHK', 'Bathroom']:
    try:
        df[col] = pd.to_numeric(df[col])
    except ValueError as e:
        print(f"Error converting column '{col}': {e}")
        # Handle the error (e.g., investigate problematic rows)
        print(f"Problematic values in column '{col}':")
        print(df[pd.to_numeric(df[col], errors='coerce').isna()])

# 3. Re-examine 'Rent' after handling
plt.figure(figsize=(8, 6))
plt.boxplot(df['Rent'])
plt.title('Box Plot of Rent After Outlier Handling')
plt.ylabel('Rent')
plt.show()

print(df.info())
```

## Data wrangling

### Subtask:
One-hot encode categorical features and scale numerical features.


**Reasoning**:
One-hot encode the categorical features and scale the numerical features as instructed.



```python
from sklearn.preprocessing import StandardScaler

# Identify categorical features for one-hot encoding
categorical_cols = ['City', 'Furnishing Status', 'Tenant Preferred', 'Point of Contact']

# One-hot encode categorical features
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Identify numerical features for scaling
numerical_cols = ['Size', 'Bathroom', 'BHK']

# Scale numerical features using StandardScaler
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

display(df.head())
```

## Feature engineering

### Subtask:
Create new features from existing ones to potentially improve model performance.


**Reasoning**:
Create new features as instructed, including 'Total Rooms', interaction terms, and features derived from 'Posted On'.



```python
import pandas as pd

# Create 'Total Rooms' feature
df['Total Rooms'] = df['BHK'] + df['Bathroom']

# Create interaction term between 'Size' and 'Total Rooms'
df['Size * Total Rooms'] = df['Size'] * df['Total Rooms']

# Create polynomial features for 'Size' (degree 2)
df['Size^2'] = df['Size'] ** 2

# Convert 'Posted On' to datetime objects
df['Posted On'] = pd.to_datetime(df['Posted On'])

# Extract month and year from 'Posted On'
df['Posted Month'] = df['Posted On'].dt.month
df['Posted Year'] = df['Posted On'].dt.year

display(df.head())
```

## Data splitting

### Subtask:
Split the data into training, validation, and testing sets.


**Reasoning**:
Split the data into training, validation, and testing sets as instructed, using `train_test_split`.

## Model optimization

### Subtask:
Optimize the hyperparameters of the trained Random Forest and Gradient Boosting models using the validation set.


**Reasoning**:
Optimize the hyperparameters of the Random Forest and Gradient Boosting models using GridSearchCV on the validation set.



```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define the parameter grid for RandomForestRegressor
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Define the parameter grid for GradientBoostingRegressor
gb_param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Instantiate GridSearchCV for RandomForestRegressor
rf_grid_search = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid=rf_param_grid,
    scoring='neg_mean_squared_error',
    cv=5,
    n_jobs=-1
)

# Instantiate GridSearchCV for GradientBoostingRegressor
gb_grid_search = GridSearchCV(
    GradientBoostingRegressor(random_state=42),
    param_grid=gb_param_grid,
    scoring='neg_mean_squared_error',
    cv=5,
    n_jobs=-1
)

# Fit GridSearchCV to the validation set
rf_grid_search.fit(X_val, y_val)
gb_grid_search.fit(X_val, y_val)

# Get the best estimators
best_rf_model = rf_grid_search.best_estimator_
best_gb_model = gb_grid_search.best_estimator_

# Print the best hyperparameters for each model
print("Best hyperparameters for RandomForestRegressor:", rf_grid_search.best_params_)
print("Best hyperparameters for GradientBoostingRegressor:", gb_grid_search.best_params_)
```

## Model evaluation

### Subtask:
Evaluate the performance of the optimized Random Forest and Gradient Boosting models on the test set.


**Reasoning**:
Evaluate the performance of the optimized Random Forest and Gradient Boosting models on the test set using appropriate metrics and compare their performance.



```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Predict on the test set
rf_predictions = best_rf_model.predict(X_test)
gb_predictions = best_gb_model.predict(X_test)

# Evaluate the models
def evaluate_model(y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    return r2, mae, rmse

rf_r2, rf_mae, rf_rmse = evaluate_model(y_test, rf_predictions)
gb_r2, gb_mae, gb_rmse = evaluate_model(y_test, gb_predictions)

print("Random Forest Model:")
print(f"R-squared: {rf_r2}")
print(f"Mean Absolute Error: {rf_mae}")
print(f"Root Mean Squared Error: {rf_rmse}")

print("\nGradient Boosting Model:")
print(f"R-squared: {gb_r2}")
print(f"Mean Absolute Error: {gb_mae}")
print(f"Root Mean Squared Error: {gb_rmse}")

# Compare the models
print("\nModel Comparison:")
if rf_r2 > gb_r2:
    print("Random Forest performs better based on R-squared.")
else:
    print("Gradient Boosting performs better based on R-squared.")

if rf_mae < gb_mae:
    print("Random Forest performs better based on MAE.")
else:
    print("Gradient Boosting performs better based on MAE.")

if rf_rmse < gb_rmse:
    print("Random Forest performs better based on RMSE.")
else:
    print("Gradient Boosting performs better based on RMSE.")
```
